## Weather Forecast Analysis

**The weather dataset**

The dataset encompasses a range of weather parameters, including temperature, humidity, wind speed, and more, providing a comprehensive overview of weather conditions across the country.

**Dataset Description** 
- country - Country code (e.g., NG - Nigeria) 
- city - City name.
- latitude - Latitude of the location. 
- longitude - Longitude of the location. 
- temp - Temperature in Kelvin.
- temp_min - Minimum temperature at the moment. 
- temp_max - Maximum temperature at the moment. 
- pressure - Atmospheric pressure on the sea level, hPa. 
- humidity - Humidity, %.
- sea_level - Atmospheric pressure on the sea level, hPa. 
- ground_level - Atmospheric pressure on the ground level, hPa. 
- wind_speed - Wind speed in meter/sec.
- wind_degree - Wind direction, degrees.
- sunrise - Sunrise time, unix, UTC.
- sunset - Sunset time, unix, UTC.
- rain_1h - Rain volume for the last 1 hour, mm.
- rain_3h - Rain volume for the last 3 hours, mm.
- timezone - Shift in seconds from UTC.
- cloud - Cloudiness, %.
- description - Weather condition within the group.
- region - State in which the city belongs.
- population - Estimated population in an area.

### Q11). What is the mean value of temp, pressure and humidity column against each 'weather condition"

```python
def cal_mean():
    categories = Weather_df['weather_condition'].unique()
    columns = ['temp', 'pressure', 'humidity']
    
    for col in columns:
        for cat in categories:
            mean = Weather_df[Weather_df['weather_condition'] == cat][col].mean()
            print(f"The mean of {col} by {cat} is: {np.round(mean,2)}")
```

```python
cal_mean()
```

### Q12). What is the minimum and maximum value of temp column against each weather condition?

```python
def temp_min_max():
    categories = Weather_df['weather_condition'].unique()
    
    for cat in categories:
        min_value = Weather_df[Weather_df['weather_condition'] == cat]['temp'].min()
        max_value = Weather_df[Weather_df['weather_condition'] == cat]['temp'].max()
        print(f"The min & max temp of by {cat} is: {min_value} and {max_value}")
```

```python
temp_min_max()
```
